Temos um Doodle com a disponibilidade de cada aluno. 

```python
import pandas as pd

df_doodle = pd.read_csv("doodle.csv")
df_doodle
```

Nosso objetivo é encontrar a quantidade minima de encontros de tal forma que todos os alunos possam participar de pelo menos um encontro.  Não precisando ser necessariamente a quantidade de encontros ótimo (ou seja, a menor possível).  Veja uma solução subótima para o problema:

```python
horarios_encontros = []

while len(df_doodle) >0:
    #desconsiderando a coluna "Aluno", procura a coluna com maior quantida de OKs
    #cada coluna representa o horario
    maior_qnt_oks = 0
    for coluna_horario in df_doodle.columns[1:]:
        qtd_oks_coluna = df_doodle[coluna_horario].count()
        if qtd_oks_coluna > maior_qnt_oks :
            maior_qnt_oks = qtd_oks_coluna
            coluna_maior_qnt_oks = coluna_horario
            
    #a coluna com maior quantidade de ok é um dos horarios de encontro
    horarios_encontros.append(coluna_maior_qnt_oks)
    
    #filtra o DataFrame colocando apenas os alunos que não foram contemplados nesse horario
    #para a proxima iteração
    df_doodle = df_doodle[df_doodle[coluna_maior_qnt_oks] != "OK"]
    
    

print(horarios_encontros)
print(f"Quantidade de encontros: {len(horarios_encontros)}")
```
